```python
marks = [30 , 55, 75, 10, 95, 60]
total = 0
highest = marks[0]
lowest = marks[0]

pass_count = 0
fail_count = 0

for mark in marks:
    total = total + mark

    if mark > highest:
        highest = mark
    if mark < lowest:
        lowest = mark
    if mark >= 40:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1
```

```python
print("Total:", total)
print("Average:", total /len(marks))
print("Highest:", highest)
print("Lowest:", lowest)
print("Pass:", pass_count)
print("Fail:", fail_count)
```

```python
marks = [85, 72, 91, 65, 40]

total = 0

for mark in marks:
    total = total + mark

average = total / len(marks)

if average >= 80:
    grade = "A"
elif average >= 70:
    grade = "B"
elif average >= 60:
    grade = "C"
elif average >= 40:
    grade = "D"
else:
    grade = "Fail"

print("------ STUDENT REPORT ------")
print("Marks:", marks)
print("Total:", total)
print("Average:", average)
print("Grade:", grade)
```
