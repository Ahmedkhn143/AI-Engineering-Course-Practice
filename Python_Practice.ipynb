```python
marks = [30 , 55, 75, 10, 95, 60]
total = 0
highest = marks[0]
lowest = marks[0]

pass_count = 0
fail_count = 0

for mark in marks:
    total = total + mark

    if mark > highest:
        highest = mark
    if mark < lowest:
        lowest = mark
    if mark >= 40:
        pass_count = pass_count + 1
    else:
        fail_count = fail_count + 1
```

```python
print("Total:", total)
print("Average:", total /len(marks))
print("Highest:", highest)
print("Lowest:", lowest)
print("Pass:", pass_count)
print("Fail:", fail_count)
```

```python
numbers = [12, 5, 18, 7, 25, 3]

# Initialize largest and smallest with the first element of the list
# This handles cases where all numbers are negative or positive correctly
largest = numbers[0]
smallest = numbers[0]

for num in numbers:
    if num > largest:
        largest = num
    if num < smallest:
        smallest = num

print("Largest number is:", largest)
print("Smallest number is:", smallest)
```

```python
# Odd numbers count karo.
# Odd numbers ka sum nikalo.
# Largest odd number find karo.
numbers = [10, 25, 8, 31, 14, 7]
odd_count = 0
odd_sum = 0
largest_odd = None # Initialize to None to correctly find the first largest odd number

for num in numbers:
    if num % 2 != 0: # Check if the number is odd
        odd_count = odd_count + 1
        odd_sum = odd_sum + num
        if largest_odd is None or num > largest_odd:
            largest_odd = num

print("Odd Count is:", odd_count)
print("Odd Sum is:", odd_sum)
print("Largest Odd is:", largest_odd)
```
